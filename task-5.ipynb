```python
# Social Media - Post Engagement Analyzer
def analyze_post_engagement(likes_list):
    total_likes = 0
    for like in likes_list:
        total_likes += like
    # Rule: If Total Likes >= 1000 -> Viral Post
    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")
analyze_post_engagement([400, 300, 350])
```

```python
# Agriculture - Rainfall Adequacy Checker
def rainfall_checker(rainfall_data, required_level):
    total = 0
    for rain in rainfall_data:
        total += rain
    average = total / len(rainfall_data)
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    print("Average Rainfall:", average)
    print("Rainfall Status:", status)
# Example
rainfall_checker([60, 70, 80, 78], 70)
```

```python
# Social Media - Duplicate Account Detection
def detect_duplicate_accounts(usernames):
    # Rule: Use set data structure
    # If the set length is less than list length, duplicates exist
    if len(usernames) != len(set(usernames)):
        found = "Yes"
    else:
        found = "No"
    print(f"Duplicate Accounts Found: {found}")
detect_duplicate_accounts(["user_A", "user_B", "user_A"])
```

```python
# System Monitoring - Application Health Checker
def evaluate_app_health(error_count):
    # Rules: 0 = Healthy, <=5 = Minor, >5 = Critical
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")
evaluate_app_health(7)
```

```python
# E-Learning – Student Attendance Eligibility System
def attendance_check(attended, total_classes):
    percentage = (attended / total_classes) * 100
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)
# Example
attendance_check(40, 50)
     
```

```python
# Smart Electricity Bill Analyzer
def calculate_smart_bill(units):
    bill = 0
    # Billing Rules: Slabs of 3, 5, and 7
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7
    if bill < 500:
        usage_status = "Low Usage"
    elif 500 <= bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"
    return bill, usage_status
total_bill, level = calculate_smart_bill(250)
print(f"Total Bill: ₹{total_bill}")
print(f"Usage Level: {level}")
```
