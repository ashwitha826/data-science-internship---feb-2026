```python
# Mobile Recharge Validation System
def validate_recharge():
    valid_plans = [199, 299, 499, 599]
    while True:  # Retry loop
        amount = int(input("Enter recharge amount: ₹"))
        # Minimum amount check
        if amount < 50:
            print("Recharge amount must be at least ₹50. Try again.\n")
            continue
        # Valid plan check
        if amount in valid_plans:
            print("Recharge Successful with plan ₹", amount)
            break
        else:
            print("Invalid plan. Valid plans are:", valid_plans)
            print("Please retry.\n")
validate_recharge()
```

```python
# Inventory Reorder Alert System
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(product, "= Reorder Alert")
        else:
            print(product, "= Stock OK")
# Sample dictionary
inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 8,
    "Monitor": 20
}
check_inventory(inventory)
```

```python
# Student Result Processing System
def check_result(marks):
    total = 0
    for m in marks:
        total += m
    avg = total / len(marks)
    print("Average:", avg)
    if avg >= 50:
        print("Pass")
    else:
        print("Fail")
marks = list(map(int, input("Enter marks separated by space: ").split()))
check_result(marks)
```

```python
# Cab Fare Estimator with Retry Option
def calculate_fare():
    while True:
        distance = float(input("Enter distance (km): "))
        if distance <= 0:
            print("Invalid distance. Try again.")
            continue
        peak = input("Peak hour? (yes/no): ").lower()
        fare = 50 + (distance * 12)
        if peak == "yes":
            fare += fare * 0.25
        print("Total Fare: ₹", round(fare, 2))
        break
calculate_fare()
```

```python
# Employee Attendance Eligibility Checker
def attendance_eligibility(attendance):
    present = 0
    for day in attendance:
        if day == "P":
            present += 1
    percentage = (present / len(attendance)) * 100
    print(f"\nAttendance Percentage: {percentage}%")
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
if __name__ == "__main__":
  status = attendance_eligibility(["P", "P", "A", "P", "P"])
  print("Eligibility:", status)
```

```python
# Password Strength Checker
def check_password_strength(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"
    if len(password) < 8:
        return "Weak Password  (Minimum 8 characters required)"
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True
    if has_digit and has_special:
        return "Strong Password "
    else:
        return "Weak Password  (Must contain digit and special character)"
if __name__ == "__main__":
  print(check_password_strength("Ashu@123"))
```
