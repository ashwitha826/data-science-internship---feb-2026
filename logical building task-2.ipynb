```python
#Unique Words in a Sentence
# Input sentence
sentence = "python is easy and python is powerful"
# Split sentence into words
words = sentence.split()
# Convert to set to remove duplicates
unique_words = set(words)
# Count unique words
count = len(unique_words)
# Output
print("Unique words count:", count)
print("Unique words:", unique_words)
```

```python
#Highest Salary from Employee Data
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
# Find employee with highest salary
highest_paid_employee = max(employees, key=employees.get)
# Get the highest salary amount
highest_salary = employees[highest_paid_employee]
# Display result
print("Highest Salary:", highest_paid_employee, "-", highest_salary)
```

```python
#Find Maximum and Minimum Values
numbers = [45, 22, 89, 10, 66]
# Initialize
maximum = numbers[0]
minimum = numbers[0]
# Loop through list
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num
# Output
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
#Count Products Above a Price Threshold
# List of product prices
prices = [450, 1200, 899, 1500, 300]
# Counter for products above 1000
count = 0
# Loop through each price
for price in prices:
    # Check if price is greater than 1000
    if price > 1000:
        count += 1   # Increase counter
# Display result
print("Products above 1000:", count)
```

```python
#Calculate Attendance Percentage
# Attendance list
attendance = ["P", "P", "A", "P", "P"]
# Count total number of days
total_days = len(attendance)
# Count number of present days
present_days = 0
for status in attendance:
    if status == "P":
        present_days += 1
# Calculate attendance percentage
attendance_percentage = (present_days / total_days) * 100
# Display result
print("Attendance Percentage:", attendance_percentage)
```

```python
#Remove Duplicate Phone Numbers
phone_numbers = [9876543210, 9123456789, 9876543210]
# Convert to set
unique_numbers = set(phone_numbers)
print("Unique phone numbers:", unique_numbers)
```

```python
#Count Character Frequency
# Input string
text = "python"
# Dictionary to store character counts
char_count = {}
# Loop through each character in the string
for char in text:
    # If character already exists in dictionary, increment count
    if char in char_count:
        char_count[char] += 1
    else:
        # If character appears first time, initialize count as 1
        char_count[char] = 1
# Display result
print(char_count)
```

```python
#Check if a Key Exists in a Dictionary
# Dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
# Key to check
key = "Ravi"
# Check existence
if key in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
#Average marks
marks = [70, 75, 80, 65]
s = 0
c = 0
for x in marks:
    s = s + x
    c = c + 1
avg = s / c
print("Average Marks:", avg)
```
